Encuentre todas las raíces del polinomio x4+540∗x3+109124∗x2+9781632∗x+328188672=0.

```python
import numpy as np
def f(x):
    return x**4 + 540*x**3 + 109124*x**2 + 9781632*x + 328188672

xs = np.linspace(-1000, 1000, 50)
ys = [f(x) for x in xs]
```

Con ayuda del siguiente codigo se puede observar los intervalos para obtener la raiz mas adelante con el metodo de la biseccion

```python
intervalos = []
for i in range(len(xs) - 1):
    if ys[i] * ys[i+1] < 0:
        intervalos.append((xs[i], xs[i+1]))

print("Intervalos con cambio de signo:")
for a, b in intervalos:
    print(f"({a}, {b})")
```

# Respuesta utilizando el metodo de la biseccion.

```python
def bisección(a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) >= 0:
        print("No se cumple f(a)*f(b) < 0. Elige otro intervalo.")
        return None

    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)

        if abs(fc) < tol or (b - a)/2 < tol:
            return c

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    print("No se encontró raíz con la tolerancia dada.")
    return None

raíz= bisección(-194.44444444444446, -138.88888888888889)
print('Raiz:', raíz)
```
