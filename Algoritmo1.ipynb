Encuentre todas las raíces del polinomio x4+540∗x3+109124∗x2+9781632∗x+328188672=0.

```python
import numpy as np
def f(x):
    return x**4 + 540*x**3 + 109124*x**2 + 9781632*x + 328188672

xs = np.linspace(-130, 125, 10)
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

fig = plt.figure()
plt.plot(xs, ys, color='white')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función f(x)')
plt.grid()
#plt.axhline(0, color='black', lw=0.5, ls='--')
#plt.axvline(0, color='black', lw=0.5, ls='--')

line, = plt.plot([], [], color='blue', marker='o', linestyle='-', label='f(x)')

def update(frame):
    line.set_data(xs[:frame], ys[:frame])
    return line,

ani = animation.FuncAnimation(fig, update, frames=len(xs)+1, interval=50, blit=True)
 
HTML(ani.to_jshtml())
```

```python
def bisección(a, b, tol=1e-6, max_iter=1000):
    if f(a) * f(b) >= 0:
        print("No se cumple f(a)*f(b) < 0. Elige otro intervalo.")
        return None

    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)

        if abs(fc) < tol or (b - a)/2 < tol:
            return c

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    print("No se encontró raíz con la tolerancia dada.")
    return None

raíz= bisección(0.0078, -0.007)
print('Raiz:', raíz)
```
